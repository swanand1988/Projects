```python
df_by_genr=movie_df.loc[:,'CVotes10':'VotesnUS']
df_by_genr[['genre_1','genre_2','genre_3']]=movie_df[['genre_1','genre_2','genre_3']]
df_by_genr
```

```python
df_by_genr['cnt']=1
df_by_genr
```

```python
df_by_g1=df_by_genr.groupby('genre_1').aggregate(np.sum)
df_by_g1
```

```python
df_by_g2=df_by_genr.groupby('genre_2').aggregate(np.sum)
df_by_g2
```

```python
df_by_g3=df_by_genr.groupby('genre_3').aggregate(np.sum)
df_by_g3
```

```python
df_add=df_by_g1.add(df_by_g2,df_by_g3, fill_value=0)
df_add
```

```python
df_add=df_by_g1.add(df_by_g2, fill_value=0)
df_add=df_add.add(df_by_g3,fill_value=0)
df_add
```

```python
df_top10=df_add.loc[df_add['cnt']>10].head(10)
df_top10
```

```python
df_top10.iloc[:,0:-1]=df_top10.iloc[:,0:-1].divide(df_top10['cnt'],axis=0)
df_top10
```

```python
df_top10.loc[:,'VotesM':'VotesnUS']=round(df_top10.loc[:,'VotesM':'VotesnUS'],2)
df_top10[df_top10.loc[:,"CVotes10":"CVotesnUS"].columns]=df_top10[df_top10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
df_top10
```

```python
sns.barplot(x=df_top10['cnt'],y=df_top10.index)
plt.show()
```

Drama is having more number of count

```python
plt.figure(figsize=(20,30))
plt.subplot(1,2,1)
hp=sns.heatmap(df_top10[['CVotesU18M','CVotes1829M','CVotes3044M','CVotes45AM']],annot=True,cmap='coolwarm')
plt.subplot(1,2,2)

hpf=sns.heatmap(df_top10[['CVotesU18F','CVotes1829F','CVotes3044F','CVotes45AF']],annot=True,cmap='coolwarm')
plt.show()
```

Inferances
1. A few inferences that can be seen from the heatmap above is that males have voted more than females, and Sci-Fi appears to be most popular among the 18-29 age group irrespective of their gender. 
2. Genre romance has got the least number of votes among any age group of males, but there is no such pattern among the females
3. Action seems to be the more popular genre among the under 18 males, and Animation appears to be the most popular genre among under 18 females.
4. 18-29 age group seems to be most actively voting for any genre irrespective of gender
